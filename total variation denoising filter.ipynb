```python
import cv2
import numpy as np
from skimage import io, img_as_float
from skimage.restoration import denoise_tv_chambolle
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
img = img_as_float(io.imread('/content/noisy2.png', as_gray=True))

plt.hist(img.flat, bins=100, range=(0,1))
```

```python
denoise_img = denoise_tv_chambolle(img, weight=0.1, eps=0.0002, n_iter_max=200, multichannel=False)
plt.imshow(denoise_img, cmap='gray')
```
