```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread("/content/noisy2.png",0)
plt.hist(img.flat, bins=100, range=(0,255))
```

```python
ret, th = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
plt.imshow(th, cmap='gray')
```

```python
ker = np.ones((3,3), np.uint8)
erosion = cv2.erode(th, ker, iterations=1)
plt.imshow(erosion, cmap='gray')
```

```python
dil = cv2.dilate(erosion, ker, iterations=1)
plt.imshow(dil, cmap='gray')
```

```python
open = cv2.morphologyEx(th, cv2.MORPH_OPEN, ker)
plt.imshow(open, cmap='gray')
```

```python
med = cv2.medianBlur(img,3)
ret, th = cv2.threshold(med, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
plt.imshow(med, cmap='gray')
```
