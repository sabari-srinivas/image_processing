```python
from skimage import io
from matplotlib import pyplot as plt
img = io.imread("/content/testimage.jpg", as_gray=True)
from skimage.transform import rescale,resize,downscale_local_mean

resc = rescale(img, 1.0/4.0, anti_aliasing=True)
resz = resize(img, (200,150))
dwn = downscale_local_mean(img, (4,3))
plt.imshow(dwn, cmap='gray')
```

```python
from skimage.filters import roberts, sobel, scharr, prewitt
rob = roberts(img)
plt.imshow(rob, cmap='gray')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, sharex=True, sharey=True, figsize=(8,8))
ax = axes.ravel()
ax[0].imshow(img, cmap=plt.cm.gray)
ax[0].set_title("original")

ax[1].imshow(rob, cmap=plt.cm.gray)
ax[1].set_title("roberts")

sob = sobel(img)
ax[2].imshow(sob, cmap=plt.cm.gray)
ax[2].set_title("roberts")

sch = scharr(img)
ax[3].imshow(sch, cmap=plt.cm.gray)
ax[3].set_title("scharr")

for a in ax:
  a.axis('off')

plt.tight_layout()
plt.show()
```

```python
from skimage.feature import canny
can = canny(img, sigma=3)
plt.imshow(can)
```

```python
from skimage import restoration
import numpy as np
psf = np.ones((3,3))/9
dec, _ = restoration.unsupervised_wiener(img,psf)
plt.imshow(dec, cmap='gray')
```
