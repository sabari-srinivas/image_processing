```python
from skimage import io, img_as_float
import matplotlib.pyplot as plt
import numpy as np
from skimage.restoration import denoise_nl_means, estimate_sigma

img = img_as_float(io.imread("/content/testimage.jpg"))
plt.hist(img.flat, bins=100, range=(0,1))
```

```python
sig = np.mean(estimate_sigma(img,multichannel=True))
pat = dict(patch_size=5, patch_distance=6, multichannel=True)
den = denoise_nl_means(img, h=1.15 * sig, fast_mode=True, **pat)
plt.hist(den.flat, bins=100, range=(0,1))
```

```python
sig = np.mean(estimate_sigma(img,multichannel=True))
pat = dict(patch_size=5, patch_distance=6, multichannel=True)
den = denoise_nl_means(img, h=1.15 * sig, fast_mode=True, **pat)

from skimage import exposure
exp = exposure.equalize_adapthist(den)

plt.hist(exp.flat, bins=100, range=(0,1))
```

```python
sig = np.mean(estimate_sigma(img,multichannel=True))
pat = dict(patch_size=5, patch_distance=6, multichannel=True)
den = denoise_nl_means(img, h=1.15 * sig, fast_mode=True, **pat)

from skimage import exposure
exp = exposure.equalize_adapthist(den)

plt.imshow(exp, cmap='gray')
```

```python
mark = np.ones(img.shape, dtype=np.uint)
mark[(exp < 0.5) & (exp > 0.3)] = 1
mark[(exp > 0.8) & (exp < 0.99)] = 2
plt.imshow(mark)
```

```python
from skimage.segmentation import random_walker
lab = random_walker(exp, mark, beta=10, mode='bf')
plt.imshow(lab)
```
