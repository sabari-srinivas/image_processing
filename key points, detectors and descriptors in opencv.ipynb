```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread("/content/grains.jpeg")

gr = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gr = np.float32(gr)
```

```python
har = cv2.cornerHarris(gr, 2, 3, 0.04)
img[har>0.01*har.max()]=[255, 0, 0]
cv2_imshow(img)
```

```python
cor = cv2.goodFeaturesToTrack(gr, 50, 0.01, 10)
cor = np.int0(cor)
for i in cor:
  x,y = i.ravel()
  cv2.circle(img,(x,y),3,255,-1)
cv2_imshow(img)
```

```python
cor = cv2.goodFeaturesToTrack(gr, 10, 0.01, 10)
cor = np.int0(cor)
for i in cor:
  x,y = i.ravel()
  print(x,y)
  cv2.circle(img,(x,y),3,255,-1)
```

```python
det = cv2.FastFeatureDetector_create(50)
kp = det.detect(img, None)
img1 = cv2.drawKeypoints(img, kp, None, flags=0)
cv2_imshow(img1)
```

```python
img = cv2.imread("/content/grains.jpeg",0)
orb = cv2.ORB_create(50)
kp, des = orb.detectAndCompute(img,None)
img1 = cv2.drawKeypoints(img, kp, None, flags=0)
cv2_imshow(img1)
```
