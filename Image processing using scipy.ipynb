```python
from skimage import io, img_as_ubyte
from scipy import ndimage
import numpy as np
from matplotlib import pyplot as plt
img = io.imread("/content/HD-wallpaper-surya-handsome-king.jpg", as_gray=True)
print(img.shape, img.dtype)
```

```python
meang = img.mean()
max = img.max()
min = img.min()
print(meang, max, min)
```

```python
flip = np.fliplr(img)
flip1 = np.flipud(img)

plt.subplot(2,1,1)
plt.imshow(img, cmap='Greys')
plt.subplot(2,2,3)
plt.imshow(flip, cmap='Blues')
plt.subplot(2,2,4)
plt.imshow(flip1, cmap='hsv')
```

```python
rot = ndimage.rotate(img,45,reshape=False)
plt.imshow(rot)
```

```python
uni = ndimage.uniform_filter(img,size=3)
plt.imshow(uni)
```

```python
guas = ndimage.gaussian_filter(img,sigma=11)
plt.imshow(guas)
```

```python
med = ndimage.median_filter(img,3)
plt.imshow(med)
```

```python
img1 = io.imread("/content/testimage.jpg", as_gray=True)
med1 = ndimage.median_filter(img1,10)
plt.imshow(med1)
```

```python
sob = ndimage.sobel(img,axis=0)
plt.imshow(sob)
```

```python
sob1 = ndimage.sobel(img1,axis=0)
plt.imshow(sob1)
```
