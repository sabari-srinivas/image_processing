```python
import numpy
from matplotlib import pyplot as plt

def guak(size,size_y=None):
  size=int(size)
  if not size_y:
    size_y=size
  else:
    size_y=int(size_y)
  x,y=numpy.mgrid[-size:size+1, -size_y:size_y+1]
  g=numpy.exp(-(x**2/float(size)+y**2/float(size_y)))
  return g / g.sum()

guak_arr = guak(3)
print(guak_arr)
plt.imshow(guak_arr, cmap=plt.get_cmap('jet'), interpolation='nearest')
plt.colorbar()
plt.show()
```

```python
from skimage import io, img_as_float
from scipy import ndimage as nd
from matplotlib import pyplot as plt
import numpy as np

img = io.imread("/content/noisy.jpg")
gau = nd.gaussian_filter(img,sigma=3)
plt.imshow(gau)
```

```python
img1 = img_as_float(io.imread("/content/noisy.jpg"))

from skimage.restoration import denoise_nl_means, estimate_sigma
sig = np.mean(estimate_sigma(img1, multichannel=True))
nlm = denoise_nl_means(img1, h=1.15 * sig, fast_mode=True, patch_size=5, patch_distance=3, multichannel=True)
plt.imshow(nlm)
```
