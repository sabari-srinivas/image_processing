```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread("/content/noisy3.jpg", 0)
eq = cv2.equalizeHist(img)

plt.hist(eq.flat, bins=100, range=(0,255))
```

```python
plt.imshow(eq, cmap='gray')
```

```python
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
cl = clahe.apply(img)
plt.imshow(cl, cmap='gray')
```

```python
ret, th1 = cv2.threshold(cl, 190, 150, cv2.THRESH_BINARY)
ret, th2 = cv2.threshold(cl, 190, 255, cv2.THRESH_BINARY_INV)
plt.imshow(th1, cmap='gray')
```

```python
plt.imshow(th2, cmap='gray')
```

```python
ret, th2 = cv2.threshold(cl, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
plt.imshow(th2, cmap='gray')
```
